```python
import numpy as np
outlier = []
def detect_z(data):
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    for i in data:
        z = (i-mean)/std
        if (np.abs(z) > thres):
            outlier.append(i)
    print(outlier)
detect_z(df['fare'])
```
